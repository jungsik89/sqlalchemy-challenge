# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
import datetime
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
inspector = inspect(engine)
session = Session(engine)
```

```python
columns = inspector.get_columns('Measurement')
for x in columns:
    print(x['name'],x['type'])
```

```python
columns = inspector.get_columns('Station')
for x in columns:
    print(x['name'],x['type'])
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
query_date = datetime.datetime.now()
# Calculate the date 1 year ago from the last data point in the database
lastdate = session.query(Measurement.date).order_by(Measurement.date.desc()).all()
lastdate = lastdate[0]
lastdate = str(lastdate)
lastdate = lastdate.replace(',', '')
lastdate = lastdate.replace('(', '')
lastdate = lastdate.replace(')', '')
lastdate = lastdate.replace("'", '')
lastdate = lastdate.replace("", '')
lastdate = dt.datetime.strptime(lastdate, '%Y-%m-%d')
lastyear =  lastdate - dt.timedelta(days=365)
print(lastdate)
print(lastyear)

# Perform a query to retrieve the data and precipitation scores
query = session.query(Measurement.date,Measurement.prcp).filter(Measurement.date >= lastyear).\
order_by(Measurement.date.asc()).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
df = pd.DataFrame(query,columns=('date','prcp'))
df = df.set_index('date')

# Sort the dataframe by date
df.sort_values(by='date')

# Use Pandas Plotting with Matplotlib to plot the data
x = df.index
y = df['prcp']
fig, ax = plt.subplots()
ax.bar(x,y,width=2.5)
fig.set_figheight(10)
fig.set_figwidth(13)
ax.set_xlabel('date')
plt.xticks([])
plt.tight_layout()
plt.show()
```

```python
# Design a query to show how many stations are available in this dataset?
query_count = session.query(Station.station).count()
query_count
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
query_active_station = session.query(Measurement.station, func.count(Measurement.id))\
                       .group_by(Measurement.station)\
                       .order_by(func.count(Measurement.id).desc()).all()

query_active_station
```

```python
most_active_station = session.query(Measurement.station)\
                    .group_by(Measurement.station)\
                    .order_by(func.count(Measurement.id).desc()).limit(1).all()

most_active_station = most_active_station[0][0]
most_active_station
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
query_temperature = session.query(func.min(Measurement.tobs),\
                            func.max(Measurement.tobs),\
                            func.avg(Measurement.tobs)).filter(Measurement.station == most_active_station).all()

query_temperature
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
query_histogram = session.query(Measurement.date, func.avg(Measurement.tobs))\
                    .filter(Measurement.date >= lastyear, Measurement.date <= lastdate)\
                    .filter(Measurement.station == most_active_station)\
                    .group_by(Measurement.date)\
                    .order_by(Measurement.date).all()

histogram_df = pd.DataFrame(query_histogram, columns=['Date', 'Temperature']).set_index(keys='Date').sort_values(by='Date')
histogram_df.head()

histogram_df.plot.hist(bins=12,)
```

```python
x = histogram_df['Temperature']
y = histogram_df['Temperature'].count()

fig, ax = plt.subplots()
ax.hist(x,edgecolor = 'black',bins=12)
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
# print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same date

#Trip is planned for 2017-08-23 to 2017-09-07. 15 Days Duration.
trip_start = '2017-08-23'
trip_end = '2017-09-07'

prev_temp_analysis = calc_temps('2016-08-23','2017-09-07')
prev_temp_analysis[0][0]

tmin = prev_temp_analysis[0][0]
tave = prev_temp_analysis[0][1]
tmax = prev_temp_analysis[0][2]
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
y = tave
x = 1
std = tmax-tmin

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(2)
ax.bar(x,y,yerr=std)
ax.set_xticks([])
ax.set_ylim(0,150)
ax.set_ylabel('Temperature(F)')

plt.tight_layout()
plt.title('Trip Avg Temp')


plt.show()

```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

total_rainfall = session.query(Measurement.station, func.sum(Measurement.tobs), Station.name,\
                Station.latitude, Station.elevation).filter(Measurement.station == Station.station).\
                group_by(Measurement.station).filter(Measurement.date.between(trip_start, trip_end)).\
                order_by(func.sum(Measurement.tobs).desc()).all()

total_rainfall
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
